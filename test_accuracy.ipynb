```python
import pickle
import pandas as pd
base_submit = pd.read_csv("./submit_file/voting-8298.csv") #最高得分结果，作为对比数据
sub_textcnn1 = pd.read_csv("./submit/submit_TextRNN.csv")

label_path = "submit_file/label_id2cate.pkl" 
with open(label_path, 'rb') as f:
    label_id2cate = pickle.load(f)
#将标签进行转换
label_cate2id = {value: key for key, value in label_id2cate.items()}
base_submit['label'] = base_submit['categories'].map(label_cate2id)
sub_textcnn1['label'] = sub_textcnn1['categories'].map(label_cate2id)

```

```python
# 与线上差0.9
score1 = sum(base_submit['label'].values == sub_textcnn1['label'].values)/10000
print("线下：{},线上估计：{}".format(score1,score1-0.0808))
```
